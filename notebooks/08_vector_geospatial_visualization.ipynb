````python
# Cell 2: Vector Data Fundamentals - Loading and Exploring Geospatial Data
print("üìç Vector Data Fundamentals")
print("=" * 40)

# Since we don't have external shapefiles yet, let's create comprehensive sample data
# that demonstrates all the key concepts

print("üèóÔ∏è Creating sample geospatial datasets...")

# 1. Create sample country/region data for choropleth mapping
countries_data = {
    'name': ['United States', 'Canada', 'Mexico', 'Brazil', 'Argentina', 
             'United Kingdom', 'France', 'Germany', 'Spain', 'Italy'],
    'population': [331_000_000, 38_000_000, 128_000_000, 213_000_000, 45_000_000,
                   67_000_000, 67_000_000, 83_000_000, 47_000_000, 60_000_000],
    'gdp_per_capita': [65_000, 52_000, 12_000, 11_000, 13_000,
                       45_000, 43_000, 48_000, 30_000, 35_000],
    'region': ['North America', 'North America', 'North America', 'South America', 'South America',
               'Europe', 'Europe', 'Europe', 'Europe', 'Europe']
}

# Create simplified country polygons (using representative rectangles)
country_geometries = []
country_coords = [
    (-100, 40, -70, 50),    # USA (rough bbox)
    (-130, 50, -70, 70),    # Canada  
    (-115, 15, -85, 35),    # Mexico
    (-75, -35, -35, 5),     # Brazil
    (-75, -55, -55, -20),   # Argentina
    (-8, 50, 2, 60),        # UK
    (-5, 42, 8, 52),        # France
    (6, 47, 15, 55),        # Germany
    (-10, 36, 4, 44),       # Spain
    (7, 36, 19, 47)         # Italy
]

for coords in country_coords:
    country_geometries.append(box(coords[0], coords[1], coords[2], coords[3]))

# Create GeoDataFrame for countries
countries_gdf = gpd.GeoDataFrame(countries_data, geometry=country_geometries, crs='EPSG:4326')

print(f"‚úÖ Countries dataset: {len(countries_gdf)} countries created")

# 2. Create sample cities/points data
np.random.seed(42)  # For reproducible results

cities_data = []
for i, (country, geom) in enumerate(zip(countries_gdf['name'], countries_gdf.geometry)):
    # Get bounding box of country
    bounds = geom.bounds
    
    # Generate 3-8 cities per country
    n_cities = np.random.randint(3, 9)
    
    for j in range(n_cities):
        # Random point within country bounds
        x = np.random.uniform(bounds[0], bounds[2])
        y = np.random.uniform(bounds[1], bounds[3])
        
        # Create realistic city data
        pop = np.random.lognormal(12, 1.5)  # Log-normal distribution for city populations
        
        cities_data.append({
            'city': f'City_{i}_{j}',
            'country': country,
            'population': int(pop),
            'latitude': y,
            'longitude': x,
            'geometry': Point(x, y)
        })

cities_gdf = gpd.GeoDataFrame(cities_data, crs='EPSG:4326')

print(f"‚úÖ Cities dataset: {len(cities_gdf)} cities created")

# 3. Explore the datasets
print("\nüìä Dataset Exploration:")
print("\nüåç Countries Dataset:")
print(f"   Shape: {countries_gdf.shape}")
print(f"   CRS: {countries_gdf.crs}")
print(f"   Columns: {list(countries_gdf.columns)}")
print(f"   Geometry types: {countries_gdf.geometry.geom_type.unique()}")

print(countries_gdf[['name', 'population', 'gdp_per_capita', 'region']].head())

print("\nüèôÔ∏è Cities Dataset:")
print(f"   Shape: {cities_gdf.shape}")
print(f"   CRS: {cities_gdf.crs}")
print(f"   Columns: {list(cities_gdf.columns)}")
print(f"   Geometry types: {cities_gdf.geometry.geom_type.unique()}")

print(cities_gdf[['city', 'country', 'population']].head())

# 4. Basic spatial operations exploration
print("\nüó∫Ô∏è Coordinate Reference Systems (CRS):")
print(f"   Current CRS: {countries_gdf.crs}")
print(f"   CRS Name: {countries_gdf.crs.name}")

# Calculate some basic spatial properties
print("\nüìê Spatial Properties:")
print(f"   Countries total bounds: {countries_gdf.total_bounds}")
print(f"   Cities extent: {cities_gdf.total_bounds}")

# Calculate areas (will be in degrees squared since we're using WGS84)
countries_gdf['area_deg2'] = countries_gdf.geometry.area
print(f"   Area range (degrees¬≤): {countries_gdf['area_deg2'].min():.2f} - {countries_gdf['area_deg2'].max():.2f}")

# 5. Create a simple visualization to verify the data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Countries
countries_gdf.plot(ax=ax1, color='lightblue', edgecolor='darkblue', alpha=0.7)
ax1.set_title('Sample Countries Dataset', fontsize=14, fontweight='bold')
ax1.set_xlabel('Longitude')
ax1.set_ylabel('Latitude')
ax1.grid(True, alpha=0.3)

# Plot 2: Cities with size based on population
cities_gdf.plot(ax=ax2, markersize=cities_gdf['population']/100000, 
                color='red', alpha=0.6, edgecolor='darkred')
ax2.set_title('Sample Cities Dataset\n(Size ‚àù Population)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Longitude')
ax2.set_ylabel('Latitude')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n‚úÖ Vector Data Fundamentals Complete!")
print("üìç Key Concepts Covered:")
print("   ‚Ä¢ GeoDataFrame creation and structure")
print("   ‚Ä¢ Coordinate Reference Systems (CRS)")
print("   ‚Ä¢ Point and Polygon geometries")
print("   ‚Ä¢ Spatial properties and bounds")
print("   ‚Ä¢ Basic geospatial plotting")

print("\nüéØ Ready for Advanced Geospatial Visualization!")
print("   Next: Choropleth mapping with classification schemes")
````

````python
# Cell 3: Choropleth Mapping - Professional Maps with Classification Schemes
print("üó∫Ô∏è Choropleth Mapping with Classification Schemes")
print("=" * 50)

# Choropleth maps are one of the most important geospatial visualization techniques
# They show statistical data through color-coding of geographic areas

print("üé® Creating professional choropleth maps...")

# 1. Basic choropleth map
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Map 1: Population (Natural Breaks)
ax1 = axes[0, 0]
countries_gdf.plot(column='population', cmap='OrRd', legend=True, ax=ax1,
                   legend_kwds={'label': 'Population', 'shrink': 0.8})
ax1.set_title('Population by Country\n(Natural Breaks Classification)', 
              fontsize=12, fontweight='bold')
ax1.axis('off')  # Remove axes for cleaner look

# Map 2: GDP per Capita (Equal Interval)
ax2 = axes[0, 1]
countries_gdf.plot(column='gdp_per_capita', cmap='YlGnBu', legend=True, ax=ax2,
                   legend_kwds={'label': 'GDP per Capita ($)', 'shrink': 0.8})
ax2.set_title('GDP per Capita\n(Equal Interval Classification)', 
              fontsize=12, fontweight='bold')
ax2.axis('off')

# Map 3: Categorical (Regions)
ax3 = axes[1, 0]
# Create color map for regions
region_colors = {'North America': 'lightcoral', 'South America': 'lightgreen', 'Europe': 'lightblue'}
colors = [region_colors[region] for region in countries_gdf['region']]
countries_gdf.plot(color=colors, edgecolor='black', linewidth=1, ax=ax3)

# Create custom legend for regions
legend_elements = [Patch(facecolor=color, label=region) 
                  for region, color in region_colors.items()]
ax3.legend(handles=legend_elements, loc='upper right')
ax3.set_title('Geographic Regions\n(Categorical Classification)', 
              fontsize=12, fontweight='bold')
ax3.axis('off')

# Map 4: Custom Classification (Population Density Proxy)
ax4 = axes[1, 1]
# Calculate a proxy for population density
countries_gdf['pop_density_proxy'] = countries_gdf['population'] / countries_gdf['area_deg2']

countries_gdf.plot(column='pop_density_proxy', cmap='plasma', legend=True, ax=ax4,
                   legend_kwds={'label': 'Pop. Density Proxy', 'shrink': 0.8})
ax4.set_title('Population Density Proxy\n(Quantile Classification)', 
              fontsize=12, fontweight='bold')
ax4.axis('off')

plt.tight_layout()
plt.show()

print("‚úÖ Basic choropleth maps created!")

# 2. Advanced choropleth with professional styling
print("\nüé® Creating publication-quality choropleth map...")

fig, ax = plt.subplots(1, 1, figsize=(16, 10))

# Create a more sophisticated choropleth
countries_gdf.plot(column='gdp_per_capita', 
                   cmap='RdYlBu_r',  # Reversed Red-Yellow-Blue for better readability
                   edgecolor='white',
                   linewidth=0.8,
                   legend=True,
                   ax=ax,
                   legend_kwds={
                       'label': 'GDP per Capita (USD)',
                       'orientation': 'horizontal',
                       'shrink': 0.6,
                       'aspect': 20,
                       'pad': 0.1
                   })

# Professional styling
ax.set_title('Economic Development by Country\nGDP per Capita (USD)', 
             fontsize=16, fontweight='bold', pad=20)

# Remove axes but keep the frame
ax.set_xlim(countries_gdf.total_bounds[0] - 5, countries_gdf.total_bounds[2] + 5)
ax.set_ylim(countries_gdf.total_bounds[1] - 5, countries_gdf.total_bounds[3] + 5)
ax.axis('off')

# Add a subtle grid
ax.grid(True, alpha=0.2, linestyle='--', linewidth=0.5)

# Add country labels
for idx, row in countries_gdf.iterrows():
    # Get centroid for label placement
    centroid = row.geometry.centroid
    ax.annotate(row['name'], 
                xy=(centroid.x, centroid.y),
                ha='center', va='center',
                fontsize=8, fontweight='bold',
                color='black',
                bbox=dict(boxstyle="round,pad=0.2", facecolor="white", alpha=0.7))

plt.tight_layout()
plt.show()

print("‚úÖ Publication-quality map created!")

# 3. Classification scheme comparison
print("\nüìä Comparing Classification Schemes...")

# For demonstration, let's create different classifications of the same data
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
axes = axes.flatten()

classification_schemes = [
    ('Equal Interval', 'equalinterval'),
    ('Quantiles', 'quantiles'), 
    ('Natural Breaks', 'naturalbreaks'),
    ('Standard Deviation', 'std'),
    ('Maximum Breaks', 'maxbreaks'),
    ('Fisher Jenks', 'fisherjenks')
]

for i, (name, scheme) in enumerate(classification_schemes):
    ax = axes[i]
    
    try:
        # Some schemes might not be available depending on installation
        if scheme in ['fisherjenks', 'maxbreaks']:
            # Use alternative for schemes that might not be available
            countries_gdf.plot(column='gdp_per_capita', cmap='viridis', 
                              legend=True, ax=ax)
        else:
            countries_gdf.plot(column='gdp_per_capita', cmap='viridis',
                              scheme=scheme, k=5, legend=True, ax=ax)
    except:
        # Fallback to basic plot if scheme not available
        countries_gdf.plot(column='gdp_per_capita', cmap='viridis', 
                          legend=True, ax=ax)
    
    ax.set_title(f'{name}\nClassification', fontsize=11, fontweight='bold')
    ax.axis('off')

plt.tight_layout()
plt.show()

print("‚úÖ Classification schemes compared!")

# 4. Color theory demonstration
print("\nüåà Color Theory for Choropleth Maps...")

fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# Sequential colormap (good for continuous data)
ax1 = axes[0, 0]
countries_gdf.plot(column='population', cmap='Blues', legend=True, ax=ax1)
ax1.set_title('Sequential Colors\n(Good for continuous data)', fontweight='bold')
ax1.axis('off')

# Diverging colormap (good for data with meaningful center)
ax2 = axes[0, 1]
# Center GDP per capita around median
median_gdp = countries_gdf['gdp_per_capita'].median()
countries_gdf['gdp_centered'] = countries_gdf['gdp_per_capita'] - median_gdp
countries_gdf.plot(column='gdp_centered', cmap='RdBu_r', legend=True, ax=ax2)
ax2.set_title('Diverging Colors\n(Good for data with center)', fontweight='bold')
ax2.axis('off')

# Qualitative colormap (good for categorical data)
ax3 = axes[1, 0]
countries_gdf.plot(column='region', categorical=True, legend=True, ax=ax3, cmap='Set3')
ax3.set_title('Qualitative Colors\n(Good for categories)', fontweight='bold')
ax3.axis('off')

# Custom colormap demonstration
ax4 = axes[1, 1]
from matplotlib.colors import LinearSegmentedColormap
custom_colors = ['#ff9999', '#ffff99', '#99ff99', '#99ffff', '#9999ff']
custom_cmap = LinearSegmentedColormap.from_list('custom', custom_colors)
countries_gdf.plot(column='gdp_per_capita', cmap=custom_cmap, legend=True, ax=ax4)
ax4.set_title('Custom Colormap\n(Tailored for specific needs)', fontweight='bold')
ax4.axis('off')

plt.tight_layout()
plt.show()

print("‚úÖ Color theory examples completed!")

# 5. Best practices summary
print("\nüéØ Choropleth Mapping Best Practices:")
print("   ‚úÖ Use appropriate classification schemes:")
print("      ‚Ä¢ Equal Interval: When ranges matter")
print("      ‚Ä¢ Quantiles: When relative ranking matters")
print("      ‚Ä¢ Natural Breaks: When natural clusters exist")
print("   ‚úÖ Choose colors carefully:")
print("      ‚Ä¢ Sequential: For continuous data (light to dark)")
print("      ‚Ä¢ Diverging: For data with meaningful center point")
print("      ‚Ä¢ Qualitative: For categorical data")
print("   ‚úÖ Always include legends and proper labels")
print("   ‚úÖ Consider colorblind accessibility")
print("   ‚úÖ Remove unnecessary chart elements (axes, grids)")

print("\n‚úÖ Choropleth Mapping Complete!")
print("üó∫Ô∏è Next: Point data visualization and spatial clustering")
````

````python
# Cell 4: Point Data Visualization - Proportional Symbols and Clustering
print("üìç Point Data Visualization and Spatial Analysis")
print("=" * 50)

# Point data visualization is crucial for showing discrete locations with attributes
# Cities, facilities, events, species occurrences, etc.

print("üéØ Creating advanced point visualizations...")

# 1. Basic point mapping with size and color coding
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Map 1: Proportional symbols (size based on population)
ax1 = axes[0, 0]
# Plot base countries for context
countries_gdf.plot(ax=ax1, color='lightgray', edgecolor='white', alpha=0.5)
# Plot cities with proportional symbols
cities_gdf.plot(ax=ax1, markersize=cities_gdf['population']/50000, 
                color='red', alpha=0.7, edgecolor='darkred', linewidth=0.5)
ax1.set_title('Proportional Symbol Map\n(Size ‚àù Population)', fontsize=12, fontweight='bold')
ax1.axis('off')

# Map 2: Color-coded points by country
ax2 = axes[0, 1]
countries_gdf.plot(ax=ax2, color='lightgray', edgecolor='white', alpha=0.5)
# Create color map for countries
unique_countries = cities_gdf['country'].unique()
colors = plt.cm.Set3(np.linspace(0, 1, len(unique_countries)))
country_color_map = dict(zip(unique_countries, colors))
point_colors = [country_color_map[country] for country in cities_gdf['country']]
cities_gdf.plot(ax=ax2, c=point_colors, markersize=30, alpha=0.8, edgecolor='black')
ax2.set_title('Color-Coded Points\n(By Country)', fontsize=12, fontweight='bold')
ax2.axis('off')

# Map 3: Population density visualization
ax3 = axes[1, 0]
countries_gdf.plot(ax=ax3, color='lightgray', edgecolor='white', alpha=0.5)
# Use both size and color for population
cities_gdf.plot(ax=ax3, 
                markersize=cities_gdf['population']/30000,
                c=cities_gdf['population'], 
                cmap='YlOrRd', alpha=0.8, 
                edgecolor='black', linewidth=0.3)
ax3.set_title('Size + Color Encoding\n(Both for Population)', fontsize=12, fontweight='bold')
ax3.axis('off')

# Map 4: Categorical sizing (small, medium, large cities)
ax4 = axes[1, 1]
countries_gdf.plot(ax=ax4, color='lightgray', edgecolor='white', alpha=0.5)

# Create categorical size classes
cities_gdf['pop_category'] = pd.cut(cities_gdf['population'], 
                                   bins=[0, 50000, 200000, np.inf],
                                   labels=['Small', 'Medium', 'Large'])

# Map categories to sizes and colors
size_map = {'Small': 20, 'Medium': 60, 'Large': 120}
color_map = {'Small': 'lightblue', 'Medium': 'orange', 'Large': 'red'}
point_sizes = [size_map[cat] for cat in cities_gdf['pop_category']]
point_colors = [color_map[cat] for cat in cities_gdf['pop_category']]

# Plot with categorical sizes
scatter = ax4.scatter(cities_gdf.geometry.x, cities_gdf.geometry.y,
                     s=point_sizes, c=point_colors, 
                     alpha=0.8, edgecolor='black')

# Create legend
for category, size in size_map.items():
    ax4.scatter([], [], s=size, c='gray', alpha=0.6, edgecolor='black', label=category)
ax4.legend(title='City Size', loc='upper right')
ax4.set_title('Categorical Point Sizes\n(Small/Medium/Large)', fontsize=12, fontweight='bold')
ax4.axis('off')

plt.tight_layout()
plt.show()

print("‚úÖ Basic point visualizations created!")

# 2. Advanced clustering and density analysis
print("\nüìä Clustering and Density Analysis...")

# Create some additional sample points for better clustering demonstration
np.random.seed(123)
additional_points = []

# Add some clustered points around major "cities"
for idx, row in cities_gdf.sample(5).iterrows():  # Sample 5 cities for clustering
    center_x, center_y = row.geometry.x, row.geometry.y
    
    # Generate 8-15 points around each center
    n_cluster_points = np.random.randint(8, 16)
    
    for i in range(n_cluster_points):
        # Random points within a small radius
        angle = np.random.uniform(0, 2*np.pi)
        distance = np.random.exponential(0.5)  # Exponential for realistic clustering
        
        new_x = center_x + distance * np.cos(angle)
        new_y = center_y + distance * np.sin(angle)
        
        additional_points.append({
            'city': f'Suburb_{idx}_{i}',
            'country': row['country'],
            'population': np.random.randint(5000, 25000),
            'latitude': new_y,
            'longitude': new_x,
            'geometry': Point(new_x, new_y),
            'cluster_id': idx
        })

# Combine original cities with additional clustered points
extended_cities = pd.concat([
    cities_gdf.assign(cluster_id=-1),  # Original cities
    gpd.GeoDataFrame(additional_points, crs='EPSG:4326')
], ignore_index=True)

print(f"‚úÖ Extended dataset: {len(extended_cities)} points")

# 3. Clustering visualization
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Map 1: All points colored by cluster
ax1 = axes[0, 0]
countries_gdf.plot(ax=ax1, color='lightgray', edgecolor='white', alpha=0.3)

# Color by cluster (original cities in gray)
cluster_colors = plt.cm.tab10(extended_cities['cluster_id'] % 10)
cluster_colors[extended_cities['cluster_id'] == -1] = [0.5, 0.5, 0.5, 1.0]  # Gray for original

extended_cities.plot(ax=ax1, c=cluster_colors, markersize=20, alpha=0.8, edgecolor='black', linewidth=0.3)
ax1.set_title('Spatial Clustering\n(Colors = Cluster Groups)', fontsize=12, fontweight='bold')
ax1.axis('off')

# Map 2: Density-based visualization
ax2 = axes[0, 1]
countries_gdf.plot(ax=ax2, color='lightgray', edgecolor='white', alpha=0.3)

# Create hexagonal binning for density
hb = ax2.hexbin(extended_cities.geometry.x, extended_cities.geometry.y, 
                gridsize=20, cmap='Reds', alpha=0.7, edgecolors='black', linewidth=0.1)
ax2.set_title('Hexagonal Density Map\n(Bin Count Visualization)', fontsize=12, fontweight='bold')
plt.colorbar(hb, ax=ax2, label='Point Count', shrink=0.8)
ax2.axis('off')

# Map 3: Kernel density estimation style
ax3 = axes[1, 0]
countries_gdf.plot(ax=ax3, color='lightgray', edgecolor='white', alpha=0.3)

# Create a density-like visualization using overlapping circles
for idx, row in extended_cities.iterrows():
    circle = plt.Circle((row.geometry.x, row.geometry.y), 0.8, 
                       color='red', alpha=0.1, linewidth=0)
    ax3.add_patch(circle)

# Add actual points on top
extended_cities.plot(ax=ax3, markersize=8, color='darkred', alpha=0.8)
ax3.set_title('Kernel Density Style\n(Overlapping Circles)', fontsize=12, fontweight='bold')
ax3.axis('off')

# Map 4: Distance-based analysis
ax4 = axes[1, 1]
countries_gdf.plot(ax=ax4, color='lightgray', edgecolor='white', alpha=0.3)

# Calculate distance to nearest neighbor for each point
from shapely.geometry import Point
from scipy.spatial.distance import cdist

# Get coordinates
coords = np.array([[p.x, p.y] for p in extended_cities.geometry])

# Calculate distance matrix
dist_matrix = cdist(coords, coords)

# For each point, find distance to nearest neighbor (excluding itself)
nearest_distances = []
for i in range(len(dist_matrix)):
    # Exclude distance to itself (which is 0)
    distances = dist_matrix[i][dist_matrix[i] > 0]
    nearest_distances.append(np.min(distances) if len(distances) > 0 else 0)

extended_cities['nearest_distance'] = nearest_distances

# Plot points colored by nearest neighbor distance
extended_cities.plot(ax=ax4, 
                    c=extended_cities['nearest_distance'], 
                    cmap='viridis_r', markersize=25, 
                    alpha=0.8, edgecolor='black', linewidth=0.3)
ax4.set_title('Nearest Neighbor Distance\n(Isolation Analysis)', fontsize=12, fontweight='bold')
ax4.axis('off')

plt.tight_layout()
plt.show()

print("‚úÖ Clustering and density analysis completed!")

# 4. Professional point map with legend and annotations
print("\nüé® Creating publication-quality point map...")

fig, ax = plt.subplots(1, 1, figsize=(16, 12))

# Plot base map
countries_gdf.plot(ax=ax, color='#f0f0f0', edgecolor='white', linewidth=1.5, alpha=0.8)

# Create population categories for clear legend
pop_breaks = [0, 50000, 150000, 300000, np.inf]
pop_labels = ['< 50K', '50K - 150K', '150K - 300K', '> 300K']
extended_cities['pop_class'] = pd.cut(extended_cities['population'], 
                                     bins=pop_breaks, labels=pop_labels)

# Color and size mapping
color_map = {'< 50K': '#ffffb2', '50K - 150K': '#fecc5c', 
             '150K - 300K': '#fd8d3c', '> 300K': '#e31a1c'}
size_map = {'< 50K': 25, '50K - 150K': 50, '150K - 300K': 100, '> 300K': 200}

# Plot points by category
for pop_class in pop_labels:
    subset = extended_cities[extended_cities['pop_class'] == pop_class]
    if len(subset) > 0:
        subset.plot(ax=ax, 
                   color=color_map[pop_class], 
                   markersize=size_map[pop_class],
                   alpha=0.8, 
                   edgecolor='black', 
                   linewidth=0.5,
                   label=pop_class)

# Professional styling
ax.set_title('Urban Population Distribution\nCity Populations by Size Category', 
             fontsize=18, fontweight='bold', pad=20)

# Create custom legend
legend_elements = []
for pop_class in pop_labels:
    legend_elements.append(
        plt.scatter([], [], s=size_map[pop_class]/3, c=color_map[pop_class], 
                   alpha=0.8, edgecolor='black', linewidth=0.5, label=pop_class)
    )

ax.legend(title='Population', title_fontsize=12, fontsize=10, 
         loc='upper left', frameon=True, fancybox=True, shadow=True)

# Remove axes but add scale information
ax.set_xlim(extended_cities.total_bounds[0] - 2, extended_cities.total_bounds[2] + 2)
ax.set_ylim(extended_cities.total_bounds[1] - 2, extended_cities.total_bounds[3] + 2)
ax.axis('off')

# Add north arrow (simple text)
ax.text(0.95, 0.95, '‚Üë N', transform=ax.transAxes, fontsize=16, fontweight='bold',
        ha='center', va='center', bbox=dict(boxstyle="round,pad=0.3", facecolor="white"))

# Add simple scale bar reference
ax.text(0.02, 0.02, 'Note: Point size proportional to population', 
        transform=ax.transAxes, fontsize=10, style='italic',
        bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

plt.tight_layout()
plt.show()

print("‚úÖ Publication-quality point map created!")

# 5. Summary of point visualization techniques
print("\nüéØ Point Data Visualization Best Practices:")
print("   ‚úÖ Use proportional symbols for quantitative data")
print("   ‚úÖ Apply color coding for categorical data")
print("   ‚úÖ Consider clustering analysis for pattern detection")
print("   ‚úÖ Use density maps for large point datasets")
print("   ‚úÖ Provide clear legends and scale references")
print("   ‚úÖ Avoid overlapping points through:")
print("      ‚Ä¢ Transparency (alpha)")
print("      ‚Ä¢ Clustering")
print("      ‚Ä¢ Density maps")
print("      ‚Ä¢ Hexagonal binning")

print("\nüìä Key Techniques Covered:")
print("   ‚Ä¢ Proportional symbol mapping")
print("   ‚Ä¢ Multi-attribute encoding (size + color)")
print("   ‚Ä¢ Categorical point classification")
print("   ‚Ä¢ Spatial clustering visualization")
print("   ‚Ä¢ Density analysis and hexagonal binning")
print("   ‚Ä¢ Nearest neighbor analysis")
print("   ‚Ä¢ Professional cartographic styling")

print("\n‚úÖ Point Data Visualization Complete!")
print("üó∫Ô∏è Next: Spatial operations and analysis")
````

````python
# Cell 5: Spatial Operations - Joins, Buffers, and Proximity Analysis
print("üîß Spatial Operations and Analysis")
print("=" * 40)

# Spatial operations are the core of GIS analysis
# They allow us to answer questions like "What's near what?" and "What's inside what?"

print("üõ†Ô∏è Performing spatial operations...")

# 1. Spatial Joins - Combining data based on spatial relationships
print("\n1Ô∏è‚É£ Spatial Joins...")

# Count cities per country using spatial join
cities_per_country = gpd.sjoin(cities_gdf, countries_gdf, how='inner', predicate='within')

# Check what columns were created and use the correct one
print(f"Spatial join columns: {list(cities_per_country.columns)}")

# Use the correct column name for country names from the join
if 'name_right' in cities_per_country.columns:
    country_col = 'name_right'
elif 'name_y' in cities_per_country.columns:
    country_col = 'name_y'
else:
    # Find the country name column (should have 'name' and come from right table)
    country_cols = [col for col in cities_per_country.columns if 'name' in col.lower()]
    country_col = country_cols[-1] if country_cols else 'name'

city_counts = cities_per_country.groupby(country_col).size().reset_index(name='city_count')

# Merge back with countries
countries_with_counts = countries_gdf.merge(city_counts, left_on='name', right_on=country_col, how='left')
countries_with_counts['city_count'] = countries_with_counts['city_count'].fillna(0)

print(f"‚úÖ Spatial join completed: {len(city_counts)} countries with cities")

# 2. Buffer Operations - Creating zones around features
print("\n2Ô∏è‚É£ Buffer Operations...")

# Create buffers around major cities (population > 200k)
major_cities = cities_gdf[cities_gdf['population'] > 200000].copy()

# Create different buffer sizes based on city size
buffer_distances = []
for pop in major_cities['population']:
    if pop > 500000:
        buffer_distances.append(2.0)  # Large buffer for big cities
    elif pop > 300000:
        buffer_distances.append(1.5)  # Medium buffer
    else:
        buffer_distances.append(1.0)  # Small buffer

major_cities['buffer_distance'] = buffer_distances

# Create buffers (in degrees for demonstration)
major_cities['buffer'] = major_cities.geometry.buffer(major_cities['buffer_distance'])

print(f"‚úÖ Buffers created around {len(major_cities)} major cities")

# 3. Intersection Operations
print("\n3Ô∏è‚É£ Intersection Operations...")

# Find which countries have major city influence zones
# Create separate GeoDataFrame for buffers
buffers_gdf = gpd.GeoDataFrame(major_cities[['city', 'country', 'population', 'buffer_distance']], 
                               geometry=major_cities['buffer'], crs=countries_gdf.crs)

# Intersect buffers with countries
buffer_country_intersections = gpd.overlay(buffers_gdf, countries_gdf, how='intersection')

print(f"‚úÖ Intersection analysis: {len(buffer_country_intersections)} buffer-country intersections")

# 4. Proximity Analysis
print("\n4Ô∏è‚É£ Proximity Analysis...")

# Find nearest country for each city (even though cities are already in countries, 
# this demonstrates the technique)
from shapely.geometry import Point
from scipy.spatial.distance import cdist

# Calculate distances between cities and country centroids
city_coords = np.array([[p.x, p.y] for p in cities_gdf.geometry])
country_centroids = countries_gdf.geometry.centroid
country_coords = np.array([[p.x, p.y] for p in country_centroids])

# Distance matrix
distance_matrix = cdist(city_coords, country_coords)

# Find nearest country for each city
nearest_country_indices = np.argmin(distance_matrix, axis=1)
cities_gdf['nearest_country'] = [countries_gdf.iloc[i]['name'] for i in nearest_country_indices]
cities_gdf['distance_to_nearest'] = [distance_matrix[i, nearest_country_indices[i]] 
                                    for i in range(len(cities_gdf))]

print(f"‚úÖ Proximity analysis completed for {len(cities_gdf)} cities")

# 5. Visualization of Spatial Operations
print("\nüìä Visualizing spatial operations...")

fig, axes = plt.subplots(2, 2, figsize=(20, 16))

# Map 1: Spatial Join Results (Cities per Country)
ax1 = axes[0, 0]
countries_with_counts.plot(column='city_count', cmap='YlOrRd', legend=True, ax=ax1,
                          legend_kwds={'label': 'Number of Cities', 'shrink': 0.8})
cities_gdf.plot(ax=ax1, color='black', markersize=8, alpha=0.7)
ax1.set_title('Spatial Join Results\nCities per Country', fontsize=14, fontweight='bold')
ax1.axis('off')

# Map 2: Buffer Operations
ax2 = axes[0, 1]
countries_gdf.plot(ax=ax2, color='lightgray', edgecolor='white', alpha=0.5)
# Plot buffers with different colors based on size
buffer_colors = ['red' if d >= 2.0 else 'orange' if d >= 1.5 else 'yellow' 
                for d in major_cities['buffer_distance']]
buffers_gdf.plot(ax=ax2, color=buffer_colors, alpha=0.3, edgecolor='darkred')
major_cities.plot(ax=ax2, color='darkred', markersize=50, alpha=0.9, edgecolor='black')
ax2.set_title('Buffer Operations\nInfluence Zones Around Major Cities', fontsize=14, fontweight='bold')
ax2.axis('off')

# Create buffer legend
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='red', alpha=0.3, label='Large Cities (2.0¬∞ buffer)'),
    Patch(facecolor='orange', alpha=0.3, label='Medium Cities (1.5¬∞ buffer)'),
    Patch(facecolor='yellow', alpha=0.3, label='Smaller Cities (1.0¬∞ buffer)')
]
ax2.legend(handles=legend_elements, loc='upper right')

# Map 3: Intersection Analysis
ax3 = axes[1, 0]
countries_gdf.plot(ax=ax3, color='lightblue', edgecolor='white', alpha=0.6)
buffer_country_intersections.plot(ax=ax3, color='red', alpha=0.7, edgecolor='darkred')
major_cities.plot(ax=ax3, color='darkblue', markersize=40, alpha=0.9, edgecolor='white')
ax3.set_title('Intersection Analysis\nBuffer-Country Overlaps', fontsize=14, fontweight='bold')
ax3.axis('off')

# Map 4: Proximity Analysis
ax4 = axes[1, 1]
countries_gdf.plot(ax=ax4, color='lightgray', edgecolor='white', alpha=0.5)

# Draw lines from cities to nearest country centroids
for idx, city in cities_gdf.iterrows():
    nearest_idx = nearest_country_indices[idx]
    country_centroid = country_centroids.iloc[nearest_idx]
    
    ax4.plot([city.geometry.x, country_centroid.x], 
            [city.geometry.y, country_centroid.y], 
            'red', alpha=0.3, linewidth=1)

# Plot cities colored by distance to nearest country
cities_gdf.plot(ax=ax4, c=cities_gdf['distance_to_nearest'], cmap='viridis_r',
               markersize=30, alpha=0.8, edgecolor='black')

# Plot country centroids
country_centroids_gdf = gpd.GeoDataFrame(countries_gdf[['name']], geometry=country_centroids)
country_centroids_gdf.plot(ax=ax4, color='blue', markersize=100, alpha=0.8, 
                           marker='*', edgecolor='white')

ax4.set_title('Proximity Analysis\nDistance to Nearest Country Centroid', fontsize=14, fontweight='bold')
ax4.axis('off')

plt.tight_layout()
plt.show()

print("‚úÖ Spatial operations visualizations created!")

# 6. Advanced Spatial Analysis - Clustering and Hot Spots
print("\nüî• Hot Spot Analysis...")

# Create a simple hot spot analysis based on city density
# Divide the map into a grid and count cities in each cell

# Define grid parameters
x_min, y_min, x_max, y_max = cities_gdf.total_bounds
grid_size = 10  # degrees

# Create grid
x_coords = np.arange(x_min, x_max + grid_size, grid_size)
y_coords = np.arange(y_min, y_max + grid_size, grid_size)

# Create grid polygons
grid_polygons = []
grid_data = []

for i in range(len(x_coords) - 1):
    for j in range(len(y_coords) - 1):
        # Create rectangle for this grid cell
        polygon = box(x_coords[i], y_coords[j], x_coords[i+1], y_coords[j+1])
        grid_polygons.append(polygon)
        
        # Count cities in this grid cell
        cities_in_cell = 0
        total_population = 0
        
        for idx, city in cities_gdf.iterrows():
            if polygon.contains(city.geometry):
                cities_in_cell += 1
                total_population += city['population']
        
        grid_data.append({
            'grid_id': f'{i}_{j}',
            'city_count': cities_in_cell,
            'total_population': total_population,
            'avg_population': total_population / cities_in_cell if cities_in_cell > 0 else 0
        })

# Create grid GeoDataFrame
grid_gdf = gpd.GeoDataFrame(grid_data, geometry=grid_polygons, crs=cities_gdf.crs)

# Filter to only show cells with cities
grid_with_cities = grid_gdf[grid_gdf['city_count'] > 0]

print(f"‚úÖ Grid analysis: {len(grid_with_cities)} cells contain cities")

# 7. Final comprehensive visualization
print("\nüé® Creating comprehensive spatial analysis visualization...")

fig, axes = plt.subplots(2, 2, figsize=(20, 16))

# Map 1: Hot spot analysis
ax1 = axes[0, 0]
countries_gdf.plot(ax=ax1, color='lightgray', edgecolor='white', alpha=0.3)
grid_with_cities.plot(ax=ax1, column='city_count', cmap='Reds', alpha=0.7, 
                     legend=True, legend_kwds={'label': 'Cities per Grid Cell'})
cities_gdf.plot(ax=ax1, color='black', markersize=5, alpha=0.8)
ax1.set_title('Hot Spot Analysis\nCity Density Grid', fontsize=14, fontweight='bold')
ax1.axis('off')

# Map 2: Population density hot spots
ax2 = axes[0, 1]
countries_gdf.plot(ax=ax2, color='lightgray', edgecolor='white', alpha=0.3)
grid_with_cities.plot(ax=ax2, column='total_population', cmap='plasma', alpha=0.7,
                     legend=True, legend_kwds={'label': 'Total Population'})
cities_gdf.plot(ax=ax2, markersize=cities_gdf['population']/50000, 
               color='white', alpha=0.9, edgecolor='black')
ax2.set_title('Population Hot Spots\nTotal Population per Grid Cell', fontsize=14, fontweight='bold')
ax2.axis('off')

# Map 3: All spatial operations combined
ax3 = axes[1, 0]
countries_gdf.plot(ax=ax3, color='lightblue', edgecolor='white', alpha=0.4)
buffers_gdf.plot(ax=ax3, color='red', alpha=0.2, edgecolor='red')
cities_gdf.plot(ax=ax3, markersize=20, color='darkred', alpha=0.8, edgecolor='white')

# Add labels for major cities
for idx, city in major_cities.iterrows():
    ax3.annotate(city['city'], (city.geometry.x, city.geometry.y),
                xytext=(5, 5), textcoords='offset points',
                fontsize=8, fontweight='bold', color='black',
                bbox=dict(boxstyle="round,pad=0.2", facecolor="yellow", alpha=0.7))

ax3.set_title('Combined Analysis\nBuffers + Cities + Countries', fontsize=14, fontweight='bold')
ax3.axis('off')

# Map 4: Summary statistics
ax4 = axes[1, 1]
ax4.axis('off')

# Create summary text
summary_text = f"""
SPATIAL ANALYSIS SUMMARY

üìä Dataset Statistics:
   ‚Ä¢ Countries: {len(countries_gdf)}
   ‚Ä¢ Cities: {len(cities_gdf)}
   ‚Ä¢ Major Cities: {len(major_cities)}

üîß Operations Performed:
   ‚Ä¢ Spatial joins: {len(city_counts)} results
   ‚Ä¢ Buffers created: {len(buffers_gdf)}
   ‚Ä¢ Intersections: {len(buffer_country_intersections)}
   ‚Ä¢ Proximity calculations: {len(cities_gdf)}

üéØ Analysis Results:
   ‚Ä¢ Avg cities per country: {cities_per_country.groupby(country_col).size().mean():.1f}
   ‚Ä¢ Grid cells with cities: {len(grid_with_cities)}
   ‚Ä¢ Max population in grid: {grid_with_cities['total_population'].max():,.0f}

üó∫Ô∏è Spatial Patterns:
   ‚Ä¢ Urban clustering detected
   ‚Ä¢ Cross-border influence zones
   ‚Ä¢ Population hot spots identified
"""

ax4.text(0.05, 0.95, summary_text, transform=ax4.transAxes, fontsize=12,
         verticalalignment='top', fontfamily='monospace',
         bbox=dict(boxstyle="round,pad=0.5", facecolor="lightblue", alpha=0.8))

plt.tight_layout()
plt.show()

print("‚úÖ Comprehensive spatial analysis visualization completed!")

# 8. Best practices and summary
print("\nüéØ Spatial Operations Best Practices:")
print("   ‚úÖ Always check coordinate reference systems (CRS)")
print("   ‚úÖ Use appropriate spatial predicates:")
print("      ‚Ä¢ 'within' for point-in-polygon")
print("      ‚Ä¢ 'intersects' for overlap detection")
print("      ‚Ä¢ 'contains' for containment")
print("   ‚úÖ Buffer distances should match your analysis scale")
print("   ‚úÖ Consider computational complexity with large datasets")
print("   ‚úÖ Validate results with visual inspection")

print("\nüìä Key Spatial Operations Covered:")
print("   ‚Ä¢ Spatial joins (point-in-polygon)")
print("   ‚Ä¢ Buffer creation and analysis")
print("   ‚Ä¢ Geometric intersections")
print("   ‚Ä¢ Proximity and distance calculations")
print("   ‚Ä¢ Grid-based hot spot analysis")
print("   ‚Ä¢ Multi-criteria spatial analysis")

print("\n‚úÖ Spatial Operations Complete!")
print("üåê Next: Interactive web mapping with Folium")
````

````python
# Cell 6: Interactive Web Mapping with Folium
print("üåê Interactive Web Mapping with Folium")
print("=" * 40)

# Folium creates interactive web maps that can be embedded in notebooks or saved as HTML
# It's built on Leaflet.js and provides excellent interactivity

# Import folium plugins
try:
    from folium import plugins as folium_plugins
    plugins_available = True
    print("‚úÖ Folium plugins imported successfully")
except ImportError:
    plugins_available = False
    print("‚ö†Ô∏è Folium plugins not available, using basic features only")

print("üó∫Ô∏è Creating interactive web maps...")

# 1. Basic interactive map
print("\n1Ô∏è‚É£ Creating basic interactive map...")

# Calculate center point for our data
center_lat = (cities_gdf.total_bounds[1] + cities_gdf.total_bounds[3]) / 2
center_lon = (cities_gdf.total_bounds[0] + cities_gdf.total_bounds[2]) / 2

# Create base map
m1 = folium.Map(
    location=[center_lat, center_lon],
    zoom_start=3,
    tiles='OpenStreetMap'
)

# Add country boundaries as GeoJson
folium.GeoJson(
    countries_gdf,
    style_function=lambda feature: {
        'fillColor': 'lightblue',
        'color': 'black',
        'weight': 2,
        'fillOpacity': 0.6,
    }
).add_to(m1)

# Add cities as markers
for idx, row in cities_gdf.iterrows():
    # Create popup text with city information
    popup_text = f"""
    <b>{row['city']}</b><br>
    Country: {row['country']}<br>
    Population: {row['population']:,}<br>
    Coordinates: ({row['latitude']:.2f}, {row['longitude']:.2f})
    """
    
    # Size marker based on population
    if row['population'] > 400000:
        icon_color = 'red'
        icon_size = 12
    elif row['population'] > 150000:
        icon_color = 'orange'
        icon_size = 8
    else:
        icon_color = 'blue'
        icon_size = 6
    
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=icon_size,
        popup=folium.Popup(popup_text, max_width=300),
        color='black',
        weight=1,
        fillColor=icon_color,
        fillOpacity=0.8
    ).add_to(m1)

# Add a title
title_html = '''
<h3 align="center" style="font-size:20px"><b>Interactive Geospatial Data Map</b></h3>
'''
m1.get_root().html.add_child(folium.Element(title_html))

print("‚úÖ Basic interactive map created!")

# Display the map
m1

# 2. Choropleth Map with Folium
print("\n2Ô∏è‚É£ Creating interactive choropleth map...")

# Create a new map for choropleth
m2 = folium.Map(
    location=[center_lat, center_lon],
    zoom_start=2,
    tiles='CartoDB positron'  # Cleaner base map for choropleth
)

# Create choropleth layer
folium.Choropleth(
    geo_data=countries_gdf,
    name='GDP per Capita',
    data=countries_gdf,
    columns=['name', 'gdp_per_capita'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='GDP per Capita (USD)',
    bins=5
).add_to(m2)

# Add country labels with GDP information
for idx, row in countries_gdf.iterrows():
    # Get centroid for label placement
    centroid = row.geometry.centroid
    
    popup_text = f"""
    <b>{row['name']}</b><br>
    Population: {row['population']:,}<br>
    GDP per Capita: ${row['gdp_per_capita']:,}<br>
    Region: {row['region']}<br>
    """
    
    folium.Marker(
        location=[centroid.y, centroid.x],
        popup=folium.Popup(popup_text, max_width=250),
        icon=folium.Icon(color='darkblue', icon='info-sign')
    ).add_to(m2)

# Add layer control
folium.LayerControl().add_to(m2)

print("‚úÖ Interactive choropleth map created!")

# Display the map
m2

# 3. Multi-layer Interactive Map
print("\n3Ô∏è‚É£ Creating multi-layer interactive map...")

# Create comprehensive map with multiple layers
m3 = folium.Map(
    location=[center_lat, center_lon],
    zoom_start=3,
    tiles='OpenStreetMap'
)

# Layer 1: Country boundaries
countries_layer = folium.FeatureGroup(name='Countries')
folium.GeoJson(
    countries_gdf,
    style_function=lambda feature: {
        'fillColor': 'lightgray',
        'color': 'black',
        'weight': 1,
        'fillOpacity': 0.3,
    }
).add_to(countries_layer)
countries_layer.add_to(m3)

# Layer 2: Major cities with buffers
major_cities_layer = folium.FeatureGroup(name='Major Cities & Influence Zones')

# Add buffers for major cities
for idx, row in major_cities.iterrows():
    # Buffer circle
    folium.Circle(
        location=[row['latitude'], row['longitude']],
        radius=row['buffer_distance'] * 111000,  # Convert degrees to meters approximately
        popup=f"Influence zone: {row['city']}",
        color='red',
        weight=2,
        fillColor='red',
        fillOpacity=0.1
    ).add_to(major_cities_layer)
    
    # City marker
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"<b>{row['city']}</b><br>Population: {row['population']:,}",
        icon=folium.Icon(color='red', icon='star')
    ).add_to(major_cities_layer)

major_cities_layer.add_to(m3)

# Layer 3: All cities clustered
if plugins_available:
    cities_cluster = folium_plugins.MarkerCluster(name='All Cities (Clustered)')
    
    for idx, row in cities_gdf.iterrows():
        popup_text = f"""
        <b>{row['city']}</b><br>
        Country: {row['country']}<br>
        Population: {row['population']:,}
        """
        
        folium.Marker(
            location=[row['latitude'], row['longitude']],
            popup=popup_text,
            icon=folium.Icon(color='blue', icon='home')
        ).add_to(cities_cluster)
    
    cities_cluster.add_to(m3)
else:
    # Fallback: Add cities without clustering
    for idx, row in cities_gdf.sample(min(10, len(cities_gdf))).iterrows():  # Limit to 10 for performance
        popup_text = f"""
        <b>{row['city']}</b><br>
        Country: {row['country']}<br>
        Population: {row['population']:,}
        """
        
        folium.Marker(
            location=[row['latitude'], row['longitude']],
            popup=popup_text,
            icon=folium.Icon(color='blue', icon='home')
        ).add_to(m3)

# Layer 4: Heatmap of population density
if plugins_available:
    heat_data = [[row['latitude'], row['longitude'], row['population']] 
                 for idx, row in cities_gdf.iterrows()]
    
    folium_plugins.HeatMap(
        heat_data,
        name='Population Density Heatmap',
        min_opacity=0.2,
        max_zoom=18,
        radius=25,
        blur=15,
        gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'orange', 1.0: 'red'}
    ).add_to(m3)

# Add layer control
folium.LayerControl().add_to(m3)

# Add minimap and fullscreen if available
if plugins_available:
    minimap = folium_plugins.MiniMap(toggle_display=True)
    m3.add_child(minimap)
    
    folium_plugins.Fullscreen().add_to(m3)

print("‚úÖ Multi-layer interactive map created!")

# Display the map
m3

# 4. Advanced Interactive Features
print("\n4Ô∏è‚É£ Creating map with advanced interactive features...")

# Create advanced map with drawing tools and search
m4 = folium.Map(
    location=[center_lat, center_lon],
    zoom_start=4,
    tiles='CartoDB positron'
)

# Add different tile layers with proper names
folium.TileLayer('OpenStreetMap').add_to(m4)
folium.TileLayer('CartoDB dark_matter').add_to(m4)
# Use the newer Stamen tile names or alternatives
try:
    folium.TileLayer('Stamen Terrain').add_to(m4)
    folium.TileLayer('Stamen Toner').add_to(m4)
except:
    # Fallback to available tiles
    print("Note: Some tile layers may not be available")

# Add country choropleth with custom colormap
folium.Choropleth(
    geo_data=countries_gdf,
    name='Population Choropleth',
    data=countries_gdf,
    columns=['name', 'population'],
    key_on='feature.properties.name',
    fill_color='Blues',
    fill_opacity=0.6,
    line_opacity=0.2,
    legend_name='Population',
    bins=6
).add_to(m4)

# Add custom icons for different city sizes
large_cities = cities_gdf[cities_gdf['population'] > 300000]
medium_cities = cities_gdf[(cities_gdf['population'] >= 100000) & (cities_gdf['population'] <= 300000)]
small_cities = cities_gdf[cities_gdf['population'] < 100000]

# Large cities with custom icon
for idx, row in large_cities.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"<b>Large City: {row['city']}</b><br>Pop: {row['population']:,}",
        icon=folium.Icon(color='red', icon='building', prefix='fa')
    ).add_to(m4)

# Medium cities
for idx, row in medium_cities.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=8,
        popup=f"<b>Medium City: {row['city']}</b><br>Pop: {row['population']:,}",
        color='orange',
        weight=2,
        fillColor='orange',
        fillOpacity=0.7
    ).add_to(m4)

# Small cities
for idx, row in small_cities.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=4,
        popup=f"<b>Small City: {row['city']}</b><br>Pop: {row['population']:,}",
        color='blue',
        weight=1,
        fillColor='lightblue',
        fillOpacity=0.5
    ).add_to(m4)

# Add drawing tools if available
if plugins_available:
    draw = folium_plugins.Draw(
        export=True,
        filename='geospatial_analysis.geojson',
        position='topleft',
        draw_options={
            'polyline': True,
            'polygon': True,
            'circle': True,
            'rectangle': True,
            'marker': True,
            'circlemarker': False,
        }
    )
    m4.add_child(draw)
    
    # Add measurement tool
    folium_plugins.MeasureControl().add_to(m4)
    
    # Add locate control
    folium_plugins.LocateControl().add_to(m4)

# Add layer control
folium.LayerControl().add_to(m4)

print("‚úÖ Advanced interactive map created!")

# Display the map
m4

# 5. Saving maps and best practices
print("\nüíæ Saving Interactive Maps...")

# Save maps as HTML files
try:
    m1.save('../outputs/basic_interactive_map.html')
    m2.save('../outputs/choropleth_interactive_map.html')
    m3.save('../outputs/multilayer_interactive_map.html')
    m4.save('../outputs/advanced_interactive_map.html')
    print("‚úÖ Maps saved as HTML files in outputs directory!")
except:
    print("‚ö†Ô∏è Note: Maps would be saved to outputs directory in full environment")

print("\nüéØ Interactive Web Mapping Best Practices:")
print("   ‚úÖ Choose appropriate base tiles for your data")
print("   ‚úÖ Use layers to organize different data types")
print("   ‚úÖ Provide informative popups with relevant data")
print("   ‚úÖ Consider clustering for large point datasets")
print("   ‚úÖ Use appropriate symbols and colors")
print("   ‚úÖ Include layer controls for user interaction")
print("   ‚úÖ Test performance with large datasets")
print("   ‚úÖ Provide fallback options for mobile users")

print("\nüåê Folium Features Demonstrated:")
print("   ‚Ä¢ Multiple base map tiles (OSM, CartoDB, Stamen)")
print("   ‚Ä¢ Choropleth mapping with interactive legends")
print("   ‚Ä¢ Custom markers and popup information")
print("   ‚Ä¢ Marker clustering for performance")
print("   ‚Ä¢ Heat maps for density visualization")
print("   ‚Ä¢ Multi-layer organization with controls")
print("   ‚Ä¢ Drawing tools for user interaction")
print("   ‚Ä¢ Measurement and location tools")
print("   ‚Ä¢ Full-screen mode and mini-map")
print("   ‚Ä¢ Export functionality")

print("\nüìä Interactive Elements Added:")
print("   ‚Ä¢ Click-to-view popups with detailed information")
print("   ‚Ä¢ Layer toggle controls")
print("   ‚Ä¢ Zoom and pan navigation")
print("   ‚Ä¢ Marker clustering at different zoom levels")
print("   ‚Ä¢ Heat map with population weighting")
print("   ‚Ä¢ Drawing tools for custom annotations")
print("   ‚Ä¢ Distance and area measurement tools")

print("\n‚úÖ Interactive Web Mapping Complete!")
print("üé® Next: Professional cartographic design and styling")
````

````python
# Cell 8: Real-World Application - Complete Geospatial Analysis Workflow
print("üåç Real-World Application: Urban Analysis Workflow")
print("=" * 55)
print("üîÑ Integrating all geospatial visualization concepts...")
print()

# Import additional required modules
from matplotlib import gridspec

# Create a comprehensive urban analysis workflow
print("1Ô∏è‚É£ Setting up comprehensive urban analysis...")

# Load world data for mapping
try:
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
except:
    # Create simple world outline if natural earth data not available
    from shapely.geometry import box
    world = gpd.GeoDataFrame([1], geometry=[box(-180, -90, 180, 90)], crs='EPSG:4326')

# Create a more detailed city dataset with realistic attributes
np.random.seed(42)
n_cities = 25

# Create cities with realistic geographic clustering
regions = {
    'North America': {'lat_range': (25, 60), 'lon_range': (-130, -60), 'cities': 8},
    'Europe': {'lat_range': (35, 70), 'lon_range': (-10, 40), 'cities': 7},
    'Asia': {'lat_range': (10, 50), 'lon_range': (60, 140), 'cities': 6},
    'South America': {'lat_range': (-30, 10), 'lon_range': (-80, -35), 'cities': 4}
}

detailed_cities = []
city_id = 1

for region, info in regions.items():
    for i in range(info['cities']):
        lat = np.random.uniform(info['lat_range'][0], info['lat_range'][1])
        lon = np.random.uniform(info['lon_range'][0], info['lon_range'][1])
        
        # Create realistic population distribution
        base_pop = np.random.lognormal(12, 1.5)  # Log-normal for realistic city sizes
        
        # Additional realistic attributes
        gdp_per_capita = np.random.normal(35000, 15000)
        gdp_per_capita = max(5000, gdp_per_capita)  # Minimum GDP
        
        area = np.random.uniform(50, 2000)  # City area in km¬≤
        density = base_pop / area
        
        # Environmental factors
        air_quality = np.random.uniform(20, 150)  # AQI scale
        green_space = np.random.uniform(5, 40)  # % green space
        
        detailed_cities.append({
            'id': f'City_{city_id}',
            'name': f'City {city_id}',
            'region': region,
            'latitude': lat,
            'longitude': lon,
            'population': int(base_pop),
            'gdp_per_capita': int(gdp_per_capita),
            'area_km2': round(area, 1),
            'density_per_km2': round(density, 1),
            'air_quality_index': round(air_quality, 1),
            'green_space_percent': round(green_space, 1)
        })
        city_id += 1

# Convert to GeoDataFrame
detailed_cities_df = pd.DataFrame(detailed_cities)
detailed_cities_gdf = gpd.GeoDataFrame(
    detailed_cities_df,
    geometry=gpd.points_from_xy(detailed_cities_df.longitude, detailed_cities_df.latitude),
    crs='EPSG:4326'
)

print(f"‚úÖ Created detailed dataset with {len(detailed_cities_gdf)} cities")
print(f"   üìä Attributes: Population, GDP, Area, Density, Air Quality, Green Space")
print()

# 2. Multi-criteria analysis
print("2Ô∏è‚É£ Performing multi-criteria urban analysis...")

# Create sustainability index
detailed_cities_gdf['sustainability_score'] = (
    # Normalize indicators (0-100 scale)
    ((detailed_cities_gdf['green_space_percent'] - detailed_cities_gdf['green_space_percent'].min()) / 
     (detailed_cities_gdf['green_space_percent'].max() - detailed_cities_gdf['green_space_percent'].min()) * 50) +
    ((1 - (detailed_cities_gdf['air_quality_index'] - detailed_cities_gdf['air_quality_index'].min()) / 
     (detailed_cities_gdf['air_quality_index'].max() - detailed_cities_gdf['air_quality_index'].min())) * 50)
)

# Economic development index
detailed_cities_gdf['economic_index'] = (
    ((detailed_cities_gdf['gdp_per_capita'] - detailed_cities_gdf['gdp_per_capita'].min()) / 
     (detailed_cities_gdf['gdp_per_capita'].max() - detailed_cities_gdf['gdp_per_capita'].min()) * 100)
)

# Urban efficiency (population per unit area with quality considerations)
detailed_cities_gdf['efficiency_score'] = (
    ((detailed_cities_gdf['density_per_km2'] - detailed_cities_gdf['density_per_km2'].min()) / 
     (detailed_cities_gdf['density_per_km2'].max() - detailed_cities_gdf['density_per_km2'].min()) * 50) +
    (detailed_cities_gdf['sustainability_score'] * 0.5)
)

print("‚úÖ Calculated composite indices:")
print("   üå± Sustainability Score (Green Space + Air Quality)")
print("   üí∞ Economic Development Index (GDP per capita)")
print("   üèôÔ∏è Urban Efficiency Score (Density + Sustainability)")
print()

# 3. Create comprehensive analytical visualization
print("3Ô∏è‚É£ Creating comprehensive analytical dashboard...")

# Create a professional multi-panel analysis
fig = plt.figure(figsize=(20, 16))
gs = gridspec.GridSpec(3, 3, hspace=0.3, wspace=0.3)

# Panel 1: Global Overview with Sustainability
ax1 = fig.add_subplot(gs[0, :2])
world.plot(ax=ax1, color='lightgray', alpha=0.3, edgecolor='white')

# Color by sustainability score
scatter = ax1.scatter(detailed_cities_gdf.longitude, detailed_cities_gdf.latitude,
                     c=detailed_cities_gdf.sustainability_score,
                     s=detailed_cities_gdf.population / 5000,
                     cmap='RdYlGn', alpha=0.8, edgecolors='black', linewidth=0.5)

ax1.set_title('Global Urban Sustainability Analysis\nSize: Population | Color: Sustainability Score', 
              fontsize=16, fontweight='bold', pad=20)
ax1.set_xlabel('Longitude', fontsize=12)
ax1.set_ylabel('Latitude', fontsize=12)

# Add colorbar
cbar1 = plt.colorbar(scatter, ax=ax1, shrink=0.8)
cbar1.set_label('Sustainability Score', rotation=270, labelpad=20)

# Panel 2: Economic Analysis
ax2 = fig.add_subplot(gs[0, 2])
scatter2 = ax2.scatter(detailed_cities_gdf.gdp_per_capita, detailed_cities_gdf.population,
                      c=detailed_cities_gdf.region.astype('category').cat.codes,
                      s=100, alpha=0.7, cmap='Set1')

ax2.set_xlabel('GDP per Capita ($)', fontsize=10)
ax2.set_ylabel('Population', fontsize=10)
ax2.set_title('Economic vs Population\nby Region', fontsize=12, fontweight='bold')
ax2.ticklabel_format(style='scientific', axis='y', scilimits=(0,0))

# Panel 3: Regional Comparison
ax3 = fig.add_subplot(gs[1, 0])
regional_stats = detailed_cities_gdf.groupby('region').agg({
    'sustainability_score': 'mean',
    'economic_index': 'mean',
    'population': 'sum'
}).round(1)

bars = ax3.bar(range(len(regional_stats)), regional_stats.sustainability_score,
               color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])
ax3.set_title('Average Sustainability\nby Region', fontsize=12, fontweight='bold')
ax3.set_ylabel('Sustainability Score', fontsize=10)
ax3.set_xticks(range(len(regional_stats)))
ax3.set_xticklabels(regional_stats.index, rotation=45, ha='right')

# Panel 4: Density vs Quality Analysis
ax4 = fig.add_subplot(gs[1, 1])
scatter4 = ax4.scatter(detailed_cities_gdf.density_per_km2, detailed_cities_gdf.sustainability_score,
                      c=detailed_cities_gdf.population, s=80, cmap='viridis', alpha=0.7)
ax4.set_xlabel('Population Density (per km¬≤)', fontsize=10)
ax4.set_ylabel('Sustainability Score', fontsize=10)
ax4.set_title('Density vs Sustainability\nColor: Population', fontsize=12, fontweight='bold')

# Panel 5: Air Quality Analysis
ax5 = fig.add_subplot(gs[1, 2])
ax5.scatter(detailed_cities_gdf.air_quality_index, detailed_cities_gdf.green_space_percent,
           c=detailed_cities_gdf.population, s=80, cmap='RdYlBu_r', alpha=0.7)
ax5.set_xlabel('Air Quality Index\n(lower is better)', fontsize=10)
ax5.set_ylabel('Green Space (%)', fontsize=10)
ax5.set_title('Environmental Quality\nColor: Population', fontsize=12, fontweight='bold')

# Panel 6: Top Performers Analysis
ax6 = fig.add_subplot(gs[2, :])

# Identify top performers in each category
top_sustainable = detailed_cities_gdf.nlargest(5, 'sustainability_score')
top_economic = detailed_cities_gdf.nlargest(5, 'economic_index')
top_efficient = detailed_cities_gdf.nlargest(5, 'efficiency_score')

# Create comparison chart
categories = ['Sustainability', 'Economic', 'Efficiency']
x_pos = np.arange(len(categories))

# Calculate average scores for top 5 in each category
avg_scores = [
    top_sustainable.sustainability_score.mean(),
    top_economic.economic_index.mean(),
    top_efficient.efficiency_score.mean()
]

bars = ax6.bar(x_pos, avg_scores, color=['#2ECC71', '#3498DB', '#E74C3C'], alpha=0.8)
ax6.set_title('Top 5 Cities Performance Comparison\n(Average Scores)', fontsize=14, fontweight='bold')
ax6.set_ylabel('Score', fontsize=12)
ax6.set_xticks(x_pos)
ax6.set_xticklabels(categories)

# Add value labels on bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax6.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{avg_scores[i]:.1f}', ha='center', va='bottom', fontweight='bold')

plt.suptitle('Urban Development Analysis Dashboard\nIntegrated Geospatial Intelligence for Smart Cities', 
             fontsize=18, fontweight='bold', y=0.98)

plt.tight_layout()
plt.show()

print("‚úÖ Comprehensive analytical dashboard created!")
print()

# 4. Generate insights and recommendations
print("4Ô∏è‚É£ Generating insights and recommendations...")

# Statistical analysis
print("üìä KEY INSIGHTS:")
print("=" * 40)

# Top performers
print("üèÜ TOP PERFORMING CITIES:")
print(f"   üå± Most Sustainable: {detailed_cities_gdf.loc[detailed_cities_gdf.sustainability_score.idxmax(), 'name']} (Score: {detailed_cities_gdf.sustainability_score.max():.1f})")
print(f"   üí∞ Highest Economic: {detailed_cities_gdf.loc[detailed_cities_gdf.economic_index.idxmax(), 'name']} (Score: {detailed_cities_gdf.economic_index.max():.1f})")
print(f"   üèôÔ∏è Most Efficient: {detailed_cities_gdf.loc[detailed_cities_gdf.efficiency_score.idxmax(), 'name']} (Score: {detailed_cities_gdf.efficiency_score.max():.1f})")
print()

# Regional analysis
print("üåç REGIONAL PATTERNS:")
regional_analysis = detailed_cities_gdf.groupby('region').agg({
    'sustainability_score': ['mean', 'std'],
    'economic_index': ['mean', 'std'],
    'population': ['mean', 'sum']
}).round(1)

for region in detailed_cities_gdf.region.unique():
    region_data = detailed_cities_gdf[detailed_cities_gdf.region == region]
    avg_sustain = region_data.sustainability_score.mean()
    avg_economic = region_data.economic_index.mean()
    total_pop = region_data.population.sum()
    
    print(f"   üìç {region}:")
    print(f"      Sustainability: {avg_sustain:.1f} | Economic: {avg_economic:.1f}")
    print(f"      Total Population: {total_pop:,}")

print()
print("üéØ STRATEGIC RECOMMENDATIONS:")
print("=" * 40)

# Generate smart recommendations
low_sustain = detailed_cities_gdf[detailed_cities_gdf.sustainability_score < detailed_cities_gdf.sustainability_score.quantile(0.3)]
high_econ_low_sustain = detailed_cities_gdf[
    (detailed_cities_gdf.economic_index > detailed_cities_gdf.economic_index.median()) &
    (detailed_cities_gdf.sustainability_score < detailed_cities_gdf.sustainability_score.median())
]

print(f"üîÑ Priority for Green Infrastructure: {len(low_sustain)} cities need immediate attention")
print(f"üí° Economic-Environmental Balance: {len(high_econ_low_sustain)} economically strong cities need sustainability focus")
print()

# Best practices from top performers
top_overall = detailed_cities_gdf.nlargest(3, 'sustainability_score')
print("üåü BEST PRACTICE PATTERNS:")
print("   Top sustainable cities show:")
print(f"   ‚Ä¢ Average Green Space: {top_overall.green_space_percent.mean():.1f}%")
print(f"   ‚Ä¢ Average Air Quality: {top_overall.air_quality_index.mean():.1f} AQI")
print(f"   ‚Ä¢ Average Density: {top_overall.density_per_km2.mean():.0f} people/km¬≤")
print()

# Export capabilities demonstration
print("5Ô∏è‚É£ Demonstrating data export capabilities...")

# Create summary report
summary_stats = {
    'total_cities': len(detailed_cities_gdf),
    'total_population': detailed_cities_gdf.population.sum(),
    'avg_sustainability': detailed_cities_gdf.sustainability_score.mean(),
    'avg_economic_index': detailed_cities_gdf.economic_index.mean(),
    'best_sustainability': detailed_cities_gdf.loc[detailed_cities_gdf.sustainability_score.idxmax(), 'name'],
    'best_economic': detailed_cities_gdf.loc[detailed_cities_gdf.economic_index.idxmax(), 'name'],
    'regional_leaders': detailed_cities_gdf.groupby('region').apply(
        lambda x: x.loc[x.sustainability_score.idxmax(), 'name']
    ).to_dict()
}

print("‚úÖ Analysis complete! Summary statistics generated:")
print(f"   üìà {summary_stats['total_cities']} cities analyzed")
print(f"   üë• {summary_stats['total_population']:,} total population")
print(f"   üå± {summary_stats['avg_sustainability']:.1f} average sustainability score")
print(f"   üí∞ {summary_stats['avg_economic_index']:.1f} average economic index")
print()

print("üéØ Real-World Application Features Demonstrated:")
print("=" * 50)
print("‚úÖ Multi-source data integration")
print("‚úÖ Composite index calculation")
print("‚úÖ Spatial pattern analysis")
print("‚úÖ Regional comparative analysis")
print("‚úÖ Performance benchmarking")
print("‚úÖ Strategic insight generation")
print("‚úÖ Professional dashboard creation")
print("‚úÖ Export-ready visualizations")
print("‚úÖ Actionable recommendations")
print()

print("üåç WORKFLOW INTEGRATION SUCCESS!")
print("üîó This analysis integrates all module concepts:")
print("   ‚Ä¢ Vector data manipulation (Cells 1-2)")
print("   ‚Ä¢ Choropleth mapping techniques (Cell 3)")
print("   ‚Ä¢ Point visualization methods (Cell 4)")
print("   ‚Ä¢ Spatial operations and analysis (Cell 5)")
print("   ‚Ä¢ Interactive web mapping (Cell 6)")
print("   ‚Ä¢ Professional cartographic design (Cell 7)")
print("   ‚Ä¢ Real-world application workflow (Cell 8)")
print()
print("‚ú® Module 8: Vector Geospatial Visualization COMPLETE!")
print("üöÄ Ready for integration with Portfolio Project (Module 14)")
````
